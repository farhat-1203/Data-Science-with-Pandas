- Reading a CSV file:

```python
import pandas as pd 
df= pd.read_csv("FinancialSample1.csv")   # here, we've simply imported a CSV file
df
```

Reading only a limited data from the CSV file:

```python
df= pd.read_csv("FinancialSample1.csv", nrows=3)   # 'nrows' attribute displays the rows upto the provided value
df
```

```python
# 'usecols' attribute displays only that particular column, whose value is provided in the brackets [] 
df= pd.read_csv("FinancialSample1.csv", usecols=[3])    
df
```

Skipping the rows in the Data Frame:

```python
df= pd.read_csv("FinancialSample1.csv", skiprows=3)  # the 'skiprows' attribute skips that no. of rows, and displays the output thereafter. (index of rows starts from 1)
df 
# The same output can be generated if we put header=3. This means we're making the row 3 as the header. 
# If we take header=None, then the column values would be taken as 0,1,2 and so on..by default, so in that case we use 'names' attribute to give the column names.
```

Setting the first column as the index column:

```python
df= pd.read_csv("FinancialSample1.csv", index_col="Segment")  # here, the first col 'Segment' is made the index column
df
```

```python
df= pd.read_csv("FinancialSample1.csv", dtype={"Units Sold":"float64"})  
df
```
