- Melt: used to transform or reshape the data frame.

```python
import pandas as pd
df= pd.read_csv("weather (2).csv")
df
```

```python
df_1=pd.melt(df, id_vars=["day"])
df_1
 # the 'melt' function takes in 2 argument; 1) the dataframe and the other id_vars=["col name to be dsiplayed as the row"].
```

```python
# to change the value of variable col name & the value col name, we use the argument var_name="___" and value_name="___"
df_1=pd.melt(df, id_vars=["day"], var_name="city", value_name="temperature")
df_1
```
