Missing values can be handled in two ways:
- The 'replace' attribute replaces a missing value with some other value as provided.
- The 'interpolate' attribute fills all the NaN values with some other value as provided, by making some kinda guesses.

```python
import pandas as pd
df= pd.read_csv("FinancialSample1.csv")
df
```

```python
new_df= df.replace(to_replace="Germany", value= "New York")  
new_df
```

```python
new_df= df.replace([],'')
```
