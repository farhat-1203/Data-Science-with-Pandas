```python
import pandas as pd
df_1=pd.DataFrame({"City": ["New York","Chicago","Orlando","Baltimore"], "Temperature":[21,14,35,32]},index=[0,1,2,3])
df_1
```

```python
df_2=pd.DataFrame({"City": ["New York","Chicago","San Francisco"], "Humidity":[68,65,71]}, index=[0,1,4])
df_2
```

```python
# the 'merge()' function joins the two data frames based on a common column.
# the 'on=__' argument takes the column name as the input on the basis of which the oprtn is to be performed.
import pandas as pd
df_1=pd.DataFrame({"City": ["New York","Chicago","Orlando","Baltimore"], "Temperature":[21,14,35,32]},index=[0,1,2,3])
df_2=pd.DataFrame({"City": ["New York","Chicago","San Francisco"], "Humidity":[68,65,71]}, index=[0,1,4])
df=pd.merge(df_1,df_2, on="City", how="inner")
df
```

```python
# the 'how=__' shows the type of join to be performed ..(inner,outer).
# the inner is the default join type, it gives only the common part as the output, which is common in both the data frames.
# the 'indicator=__' specifies the originating data frame of the element.

df=pd.merge(df_1,df_2, on="City", how="outer", indicator=True)
df
```

```python
df_3=pd.DataFrame({"City": ["New York","Chicago","Orlando","Baltimore"], "Temperature":[21,14,35,32], "Humidity":[65,68,71,75]},index=[0,1,2,3])
df_3
```

```python
df_4=pd.DataFrame({"City": ["Chicago","New York","San Diego"],"Temperature":[21,14,35], "Humidity":[68,65,71]}, index=[0,1,4])
df_4
```

```python
# the 'suffixes=[_,_]' decides the suffix to be putted, if there's a common column in between the data frames.
df=pd.merge(df_3,df_4, on="City", suffixes=["_1stcol","_2ndcol"]) 
df
```
