- PIVOT:  allows to transform or reshape the dataframe

```python
import pandas as pd
df=pd.read_csv("weather.csv")
df
```

```python
# The 'pivot' function reshapes the data frame, as per our demand.
""" The 'index' argument takes the column name as the input, which is to be treated as the x-axis (Row),
and the 'columns' argument takes the column name which is to be treated as the column. 
"""
# The 'values' argument takes the value as the input which is to be  displayed as the final output.
df.pivot(index="date", columns="city", values="temperature")
```

```python
df.pivot(index="humidity", columns="date")
```

- PIVOT TABLE: used to summarize and aggreagate the data inside the data frame.


```python
df= pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="date", columns="city")
# the 'pivot table' calculates the stats based on the given data.
# the aggfunc="mean" is the default type that it takes.
```

```python
df.pivot_table(index="date", columns="city", aggfunc="sum", margins=True)
# the 'margins=True' dsiplays the overall statistics
```

- Grouper function:
we use this function to aggreagte.

```python
df.pivot_table(index=pd.Grouper(freq="M", key="date"), columns="city", aggfunc="sum")
```
