```python
import pandas as pd
India_weather=pd.DataFrame({                      # here we simply created a data frame for the weather conditions of a few Indian cities.
    "City":["Mumbai","Delhi","Bangalore"],           
    "Temperature":[32,45,30],
    "Humidity":[80,60,78]
})
India_weather
```

```python
US_weather=pd.DataFrame({
    "City":["New York","Chicago","Orlando"],
    "Temperature":[21,14,35],
    "Humidity":[68,65,75]
})
US_weather
```

```python
# To concat means to append (join) two or more data frames.
df=pd.concat([India_weather,US_weather])
df
```

```python
# the concat function will take the previous indexes by deafult, so if we want to allot new indexes, then we use the arguement 'ignore_index=True'.
df=pd.concat([India_weather,US_weather], ignore_index=True)
df
```

```python
# the keys argument gives a key (a new name or a new index to be precise) to each data frame appended (so as to identify more clearly).
df=pd.concat([India_weather,US_weather], keys=["India","US"])
df
```

```python
df.loc["India"]   # the loc function takes in the index as the input and returns the data frame which is on that particular index.
```

- To append the Data Frames by columns, we use axis=1.
- axis=0, i.e concat by rows is default.

```python
Indian_temp=pd.DataFrame({
    "City":["Mumbai","Delhi","Bangalore"],           
    "Temperature":[32,45,30]
}, index=[0,1,2])
Indian_temp
```

```python
India_windspeed=pd.DataFrame({
    "City":["Delhi","Bangalore","Mumbai"],           
    "Windspeed":[7,12,9]    
}, index=[1,2,0])
India_windspeed
```

```python
df= pd.concat([Indian_temp,India_windspeed])   # here, we just used the concat function, so the append is by default row-wise.
df
```

```python
df=pd.concat([Indian_temp,India_windspeed], axis= 1)   # to append by columns, we used here the axis=1 parameter.
df
```
